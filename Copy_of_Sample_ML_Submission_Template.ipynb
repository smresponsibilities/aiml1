## ***5. ML Model Implementation***

### a> Data Scaling

````python
# Scaling your data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['Open', 'High', 'Low', 'Adj Close', 'Volume']
data[columns_to_scale] = standardScaler.fit_transform(data[columns_to_scale])
````

##### Which method have you used to scale you data and why?

We used standard scaling as we wanted all columns to contribute equally to the model.

### b> Data Splitting

````python
# Split your data to train and test. Choose Splitting ratio wisely.
x=data[['Open', 'High', 'Low',  'Volume']].values
y=data['Close']
````

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
x_test
````

##### What data splitting ratio have you used and why?

We split the data into 70-30 split into training and testing as we didn't wanted to cross validate.

### ML Model - 1

````python
#ML Model - 1
#Linear Regression
import math
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, RidgeCV, Lasso
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
lr = LinearRegression()
# Fit the Algorithm
lr.fit(x_train,y_train)
# Predict on the model
ytrainlr_pred=lr.predict(x_train)
ytestlr_pred=lr.predict(x_test)
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

````python
# Visualizing evaluation Metric Score chart
print("The train score for ls model is {}".format(lr.score(x_train,y_train)))
print("The train score for lr model is {}".format(lr.score(x_test,y_test)))
LR_MSE  = mean_squared_error(y_test, ytestlr_pred)
print("MSE :" , LR_MSE)
LR_RMSE = np.sqrt(LR_MSE)
print("RMSE :" ,LR_RMSE)
LR_r2 = r2_score(y_test, ytestlr_pred)
print("R2 :" ,LR_r2)
LR_adjusted_r2 = 1-(1-r2_score((y_test),(ytestlr_pred)))*((x_test.shape[0]-1)/(x_test.shape[0]-x_test.shape[1]-1))
print("Adjusted R2 : ",LR_adjusted_r2)
````
